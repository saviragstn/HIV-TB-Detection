Read Data

```python
dtrain = pd.read_csv('train.csv')
dtest = pd.read_csv('test.csv')
```

```python
x_train = dtrain.drop(['Label'], axis=1)
y_train = dtrain['Label']
```

```python
x_test = dtest.drop(['Label'], axis=1)
y_test = dtest['Label']
```

# FEATURE SELECTION + NORMALIZATION + MODELLING DATA  USING SVM DEFAULT PARAMETER

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC  
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from numpy import mean
from numpy import std
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cv = KFold(n_splits = 10, random_state=None, shuffle=False)
```

```python
seed(8)
tf.random.set_random_seed(8)
```

```python
accuracylinear_train = []
accuracylinear_test = []
accuracypoly_train = []
accuracypoly_test = []
accuracyrbf_train = []
accuracyrbf_test = []
x_train_new = []
x_test_new = []
component = np.arange(1, 30)

for i in range(len(component)):
    n = component[i]
    print('Banyaknya Fitur:', n)
    pca = PCA(n, svd_solver= "arpack", whiten=True)
    x_train_pre = pca.fit_transform(x_train)
    x_test_pre = pca.transform(x_test)
    x_train_new.append(x_train_pre)
    x_test_new.append(x_test_pre)
    print(pca)
    
    #LINEAR TRAIN
    model_lin = SVC(kernel='linear')  
    scores_train_lin = cross_val_score(model_lin, x_train_new[i], y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Train Accuracy for Linear Model Default Parameter is %.2f with a standard deviation of %.2f' % (scores_train_lin.mean(), scores_train_lin.std()))
    accuracylinear_train.append(mean(scores_train_lin))
    #LINEAR TEST 
    scores_test_lin = cross_val_score(model_lin, x_test_new[i], y_test, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Test Accuracy for Linear Model Default Parameter is %.2f with a standard deviation of %.2f' % (mean(scores_test_lin), std(scores_test_lin)))
    accuracylinear_test.append(mean(scores_test_lin))
    
    #POLY TRAIN
    model_poly = SVC(kernel='poly') 
    scores_train_poly = cross_val_score(model_poly, x_train_new[i], y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Train Accuracy for Poly Model Default Parameter is %.2f with a standard deviation of %.2f' % (mean(scores_train_poly), std(scores_train_poly)))
    accuracypoly_train.append(mean(scores_train_poly))
    #POLY TEST
    scores_test_poly = cross_val_score(model_poly, x_test_new[i], y_test, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Test Accuracy for Poly Model Default Parameter is %.2f with a standard deviation of %.2f' % (mean(scores_test_poly), std(scores_test_poly)))
    accuracypoly_test.append(mean(scores_test_poly))
    
    #RBF TRAIN
    model_rbf = SVC(kernel='rbf') 
    scores_train_rbf = cross_val_score(model_rbf, x_train_new[i], y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Train Accuracy for RBF Model Default Parameter is %.2f with a standard deviation of %.2f' % (mean(scores_train_rbf), std(scores_train_rbf)))
    accuracyrbf_train.append(mean(scores_train_rbf))
    #RBF TEST
    scores_test_rbf = cross_val_score(model_rbf, x_test_new[i], y_test, scoring='accuracy', cv=cv, n_jobs=-1)
    print('Data Test Accuracy for RBF Model Default Parameter is %.2f with a standard deviation of %.2f' % (mean(scores_test_rbf), std(scores_test_rbf)))
    accuracyrbf_test.append(mean(scores_test_rbf))
    print('')
```
