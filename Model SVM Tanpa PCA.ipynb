# READ DATA

```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from numpy.random import seed
import tensorflow as tf
import pickle

data = pd.read_csv('GSE50834_data_set.csv')
data
```

# CHECKING MISSING VALUE

```python
print('Total Missing Value pada Data HIV:',data.isnull().sum().sum())
```

```python
dict = {'H': 0, 'HT': 1} 
data.replace(dict, inplace=True)
data = data.drop(['Unnamed: 0'],axis=1)
data.rename({"Type":"Label"}, axis="columns", inplace=True)
data.head()
```

# SPLITTING DATA

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size = 0.30, random_state = 5)
test
```

# SAVING DATA TO CSV

```python
train.to_csv('train100%.csv', index=False)
test.to_csv('test100%.csv', index=False)
```

Read Data

```python
trainwopca = pd.read_csv('train.csv')
testwopca = pd.read_csv('test.csv')
```

# NORMALIZATION

```python
from sklearn.preprocessing import  MinMaxScaler

scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(trainwopca)
scaled_test = scaler.transform(testwopca)

scaled_train = DataFrame(trainwopca)
scaled_test = DataFrame(testwopca)
```

```python
x_train = trainwopca.drop(['Label'], axis=1)
y_train = trainwopca['Label']

x_test = testwopca.drop(['Label'], axis=1)
y_test = scaled_test['Label']
```

# MODEL CLASSIFICATION

```python
from sklearn.svm import SVC  
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

seed(8)
tf.random.set_random_seed(8)
```

# 1. Linear

```python
model_lin = SVC(kernel='linear') 
model_lin.fit(x_train, y_train)

prediction_linear = model_lin.predict(x_train)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_train, prediction_linear))
print('Accuracy Data Train:', accuracy_score(y_train, prediction_linear))

prediction_linear = model_lin.predict(x_test)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_test, prediction_linear))
print('Accuracy Data Test:', accuracy_score(y_test, prediction_linear))
```

# 2. Poly

```python
model_poly = SVC(kernel='poly') 
model_poly.fit(x_train, y_train) 
   
prediction_poly = model_poly.predict(x_train)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_train, prediction_poly))
print('Accuracy Data Train:', accuracy_score(y_train, prediction_poly))

prediction_poly = model_poly.predict(x_test)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_test, prediction_poly))
print('Accuracy Data Test:', accuracy_score(y_test, prediction_poly))
```

# 3. RBF

```python
model_rbf = SVC(kernel='rbf') 
model_rbf.fit(x_train, y_train) 

prediction_rbf = model_rbf.predict(x_train)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_train, prediction_rbf))
print('Accuracy Data Train:', accuracy_score(y_train, prediction_rbf))

prediction_rbf = model_rbf.predict(x_test)
print('Confusion matrix from testing data:')
print(confusion_matrix(y_test, prediction_rbf))
print('Accuracy:', accuracy_score(y_test, prediction_rbf))
```
